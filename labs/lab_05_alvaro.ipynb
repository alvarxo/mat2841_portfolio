EL conjunto de datos se denomina `company_sales_data.csv`, el cual contiene información tal como: número del mes, unidades, precio, etc.

Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","company_sales_data.csv"))
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

**Observación.-** Puedes ocupar las librerías de Matplolib o Seaborn.

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
# plot matplotlib
fig = plt.figure(figsize=(10, 5)) 
plt.plot(
    df['month_number'], # eje x
    df['total_profit'], # eje y
    color="black", # color
    linewidth=1 # tamano de la curva
)
plt.title("beneficios totales último año") # agregar titulo
plt.xlabel('mes') # nombre eje x
plt.ylabel('Beneficio Total') # nombre eje y
plt.grid() # agregar grillado
plt.show() # mostrar grafico


plt.figure(figsize=(10, 5))
sns.scatterplot(
    x='month_number',
    y='total_profit',
    data=df
)   
plt.show()
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
#NOTA: asumiré que los valores mostrados en "facecream	facewash	toothpaste	bathingsoap	shampoo	moisturizer" 
#es número de ventas y no dinero

# plot matplotlib
fig = plt.figure(figsize=(10, 5)) #genero la figura

#grafico "facecream"
plt.plot(
    df['month_number'], # eje x
    df['facecream'], # eje y
    label='facecream', # etiquetado
    color="black", # color
    linewidth=1 # tamano de la curva
)


#grafico "facewash"
plt.plot(
    df['month_number'], # eje x
    df['facewash'], # eje y
    label='facewash', # etiquetado
    color="blue", # color
    linewidth=1 # tamano de la curva
)

#grafico "thootpaste"
plt.plot(
    df['month_number'], # eje x
    df['toothpaste'], # eje y
    label='toothpaste', # etiquetado
    color="red", # color
    linewidth=1 # tamano de la curva
)

#grafico "bathingsoap"

plt.plot(
    df['month_number'], # eje x
    df['bathingsoap'], # eje y
    label='bathingsoap', # etiquetado
    color="green", # color
    linewidth=1 # tamano de la curva
)

#grafico "shampoo"

plt.plot(
    df['month_number'], # eje x
    df['shampoo'], # eje y
    label='shampoo', # etiquetado
    color="yellow", # color
    linewidth=1 # tamano de la curva
)

#grafico "moinsturizer"
plt.plot(
    df['month_number'], # eje x
    df['moisturizer'], # eje y
    label='moisturizer', # etiquetado
    color="brown", # color
    linewidth=1 # tamano de la curva
)
plt.legend() # agregar etiquetado
plt.title("ventas de productos último año") # agregar titulo
plt.xlabel('mes') # nombre eje x
plt.ylabel('Cantidad de ventas') # nombre eje y
plt.grid() # agregar grillado
plt.show() # mostrar grafico
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
# grafico
fig = plt.figure(figsize=(10, 5))

#grafico de barras "facecream"

plt.bar(
    df['month_number'], # eje x
    df['facecream'], # eje y 
    align='center', # centrar nombre eje x
    label='facecream', # etiquetado
    color="black", # color 
    alpha=0.6 # intensidad del color
)

#grafico de barras "facewash"

plt.bar(
    df['month_number'], # eje x
    df['facewash'], # eje y 
    align='center', # centrar nombre eje x
    label='facewash', # etiquetado
    color="blue", # color 
    alpha=0.6 # intensidad del color
)
plt.legend() # agregar etiquetado
plt.ylabel('ventas')
plt.xlabel('mes')
plt.show()
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
#grafico box-plot

nombres=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'] #datos que quiero graficar

plt.figure(figsize=(10, 6))
sns.boxplot(data=df[nombres])
plt.show()
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
#grafico círcular

plt.pie(df[nombres].sum(), labels=nombres, autopct="%0.1f %%") #se solicitan los datos que ya están almacenados en la variable "nombres"
plt.axis("equal")
plt.show()
```
